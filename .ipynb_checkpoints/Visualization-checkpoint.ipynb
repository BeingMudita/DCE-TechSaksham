```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.array([0,6])
y= np.array([0,50])
plt.plot(x,y)
plt.show()
```

```python
x= np.random.rand(10)
y= np.random.rand(10)
plt.plot(x,y)
plt.show()
```

```python
y= np.array([0,50])
plt.plot(y, linestyle = "dashdot")
plt.show()
```

```python
x= np.array([5,7,3,8,2,8,45,10])
y= np.array([3,7,6,34,83,67,5,15])
plt.scatter(x,y)
plt.show()
```

```python
x= np.array(["A", "B", "C", "D"])
y= np.array([90, 50, 80,20])
plt.bar(x,y)
plt.show()
```

```python
x= np.array(["A", "B", "C", "D"])
y= np.array([90, 50, 80,20])
plt.barh(x,y)
plt.show()
```

```python
x= np.array([20,50,60, 80])
label = ["Elsa", "Snow white","Bella", "Cindrella"]
plt.pie(x, labels= label)
plt.show()
```
