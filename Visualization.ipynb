```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.array([0,6])
y= np.array([0,50])
plt.plot(x,y)
plt.show()
```

```python
x= np.random.rand(10)
y= np.random.rand(10)
plt.plot(x,y)
plt.show()
```

```python
y= np.array([0,50])
plt.plot(y, linestyle = "dashdot")
plt.show()
```

```python
x= np.array([5,7,3,8,2,8,45,10])
y= np.array([3,7,6,34,83,67,5,15])
plt.scatter(x,y)
plt.show()
```
